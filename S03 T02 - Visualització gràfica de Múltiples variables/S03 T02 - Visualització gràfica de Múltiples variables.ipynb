```python
# Load movies.dat dataset and asign names for the columns

col_names = ['Id', 'Title year', 'Genres']
movies = pd.read_table("movies.dat", sep ="::", header=None, engine='python', names=col_names)
```

### Exploratory Data Analysis

First I start looking the first 5 rows to get a glimpse of the structure of the data

```python
movies.head()
```

There are only 3 columns:

1. **Id**: Movie ID
2. **Movie Year**: Title and year of the movie
3. **Genres**: Genres of the movie

I see that the year is a part of the title's string. In order to get some useful information and statistics I want to separate title and year and store them in two different columns:

```python
# Create 2 new columns to store the name of the movie and the year separately
movies[["Title", "Year"]] = movies["Title year"].str.extract(r'(.+)\(([0-9]{4})\)')
movies.head()
```

```python
# Drop the Movie year column and change the order of the columns

movies = movies.drop("Title year", axis=1)

movies = movies[["Title", "Year", "Genres"]]

movies.head()
```

I can see that the column Genres can contain more than one genre for each movie. They are stored as one single string with the genres separated by a "|". This makes difficult to get information about the genres. My idea is to create new columns for each of the genres and use **get_dummies()** to asign 1 or 0 to each column for the corresponding movie

```python
movies = movies.join(movies['Genres'].str.get_dummies('|'))
```

```python
# Show the first row of the dataset
movies.iloc[0]
```

I added a column for each of the possible genres and added 1 if the genre was asigned to the movie and 0 if it wasn't. If we take a look at the movie in the first row, Toy Story, in the column Genres we have these values __[Animation|Children's|Comedy]__. In the corresponding columns, we can see that these genres have a 1 as value.

### Two variables: Categoric and Numeric

#### 1 Categoric: Year / 1 Numeric: total_films

```python
# Group the movies per year to get the number of entries using the function size()

year_grouped = movies.groupby("Year", as_index=False)
year_grouped.size().head()
```

```python
# Figure Number of movies per year
fig, ax = plt.subplots(figsize=(18,7))

sns.set_style("whitegrid")

ax = sns.lineplot(x="Year", y="size", data = year_grouped.size())

labels = ax.get_xticklabels()
plt.setp(labels, rotation=90, horizontalalignment="right")

ax.set_title("Number of movies per year", fontsize="20", weight="bold")

ax.set_xlabel("Years", fontsize="15")

ax.set_ylabel("Number of movies", fontsize="15")

plt.show()
```

There are many years and for this reason it´s not easy to see the information. It´s better to show the number of movies for each decade in the dataset. This info should be easier to understand. I will cut the Year column in different bins to classify the years in decades and add this bins in a new column. I will use this new column to group the movies per decade 

```python
# Create a column with bins for decades
cut_bins = np.arange(1909, 2019, 10)
cut_labels = [str(x+1) for x in cut_bins[:-1]]

movies["Decades"] = pd.cut(movies['Year'].astype('int'), bins = cut_bins, labels = cut_labels)
```

```python
dec_grouped = movies.groupby("Decades", as_index=False)
dec_grouped.size().head()
```

```python
# Figure Number of movies per year
fig, ax = plt.subplots(figsize=(15,7))

ax = sns.barplot(x="Decades", y="size", data = dec_grouped.size())

labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment="right")

ax.set_title("Number of movies per decade", fontsize="20", weight="bold")

ax.set_xlabel("Decades", fontsize="15")

ax.set_ylabel("Number of movies", fontsize="15")

plt.show()
```

Now it´s easier to see the info. As we can see, the number of movies per decade increases as of 1970. It seems that the number of movies decreases in the 2000 - 2010 decade but this is because we only have data for one year of the decade, the year 2000

#### Mean of number of words of the title per genre

I want to see the mean of number of words of the title per genre. I will use the **get_num_words()** function to get the number of words of every title. I won´t count the punctuation marks. To do this, I will use the _nltk_ module to tokenize the text and a list of puntuation marks to discard them.

I will add a column with the total of words obtained using the **apply()** funtion. Finally I will create a new data set with a column for the genre and another column with the mean of the words.

```python
import nltk
import string
punct = []
punct[:0] = string.punctuation
```

```python
def get_num_words(row):
    words = nltk.word_tokenize(row["Title"].lower())
    words = [w for w in words if w not in punct]
    return len(words)
```

```python
movies["Num_words"] = movies.apply(get_num_words, axis=1)
```

```python
num_words = {"Genre" : [], "Num_words" : []}
for g in movies.columns.difference(['Title', 'Year', 'Genres','Decades', "Num_words"]).tolist():
    num_words["Genre"].append(g)
    num_words["Num_words"].append(movies[movies[g] == 1]["Num_words"].mean())
df_num_words = pd.DataFrame(num_words)
```

```python
df_num_words.head()
```

```python
plt.figure(figsize=(10,7))

ax = sns.barplot(y="Genre", x="Num_words", data=df_num_words.sort_values("Num_words", ascending=False), 
                 color="#0099ff")

ax.set_title("Mean of number of words of the title per genre", fontsize="15", weight="bold")
ax.set_xlabel("Number of words", fontsize="15")
ax.set_ylabel("Genres", fontsize="15")

plt.show()
```

#### Correlation of genres

I´ve seen that many movies have more than one gender. For this reason, I want to see if there are genres that tend to appear together. I will count the times a genre appears together with another one to see if there is a correlation.

Surprisingly, I found out that "Drama" and "Comedy" are strongly realted. 226 movies use these two words to define the genre. Both are too strongly related to "Romance".

Not so surprising, I've found out that "Action", "Adventure" and "Thriller" are used to define some movies and "Children´s" with "Animation", "Adventure" and "Comedy"

"Documentary" doesn´t correlate to other genres.

```python
import itertools
```

```python
genres = movies.columns.difference(['Title', 'Year', 'Genres','Decades', "Num_words"]).tolist()
df = pd.DataFrame(0, index=genres, columns=genres)
```

```python
movies = movies.drop('Num_words', 1)
```

```python
def get_counts(g):
    perm = list(itertools.permutations(g))
    counts = {}
    for p in perm:
        if p[0] not in counts:
            counts.update({p[0]:p[1:]})
    return counts
```

```python
for index, row in movies.iterrows():
    g = row[row == 1].index
    counts = get_counts(g)
    for k,v in counts.items():
        for value in v:
            df.loc[k][value] += 1
```

```python
df.head()
```

```python
plt.figure(figsize=(15,10))

ax = sns.heatmap(df, cmap="Greens", annot=True, square=True, fmt="d", linewidth=0.3)

ax.set_title("Correlation of genres", fontsize="15", weight="bold")
ax.set_xlabel("Genres", fontsize="15")
ax.set_ylabel("Genres", fontsize="15")

plt.show()
```
